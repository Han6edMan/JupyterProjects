## os.path.getatime()
`os.path.getatime(filename)`

返回`filename`最后一次访问时间，该数据由`os.stat()`获得

## os.path.getctime()

`os.path.getctime(filename)`

返回`filename`元数据改变的时间，该数据由`os.stat()`获得

## os.path.getmtime()
`os.path.getmtime(filename)`

返回`filename`最后一次修改的时间，该数据由`os.stat()`获得


````python
import time

def fileTime(file):
    return [time.ctime(os.path.getatime(file)),
            time.ctime(os.path.getctime(file)),
            time.ctime(os.path.getmtime(file))]

times = fileTime("c:/Windows")
print(times)
````

## os.path.pardir


**Docstring**:

str(object='') -> str
str(bytes_or_buffer[, encoding[, errors]]) -> str

也可以是`os.pardir`，即`..`字符串，从给定对象创建一个新的字符串对象。如果指定了编码方式或错误，则对象必须公开一个数据缓冲区，该缓冲区将使用给定的编码和错误处理程序进行解码；否则返回`object.__str__()`的结果（如果定义了的话），或返回`repr(object)`的结果，编码方式默认为`sys.getdefaultencoding()`，错误默认为`strict`

````python
path = os.getcwd()
print("current dir:", path)
parent = os.path.join(path, os.path.pardir)
parent = os.path.join(parent, os.pardir)
print("parent dir:", parent)
print("parent dir:", os.path.abspath(parent))
````

# os.remove()

`os.remove(path, *, dir_fd=None)`

__Docstring__

删除一个文件，与unlink()相同

若`dir_fd`不为`None`，则it should be a file descriptor open to a directory，并且路径（path）应该是相对的，即路径将相对于该directory的。有些操作平台上可能无法使用`dir_fd`，其无法使用时会抛出`NotImplementedError`的异常

**Type**:      builtin_function_or_method
