# Jupyter Notebook

## 鼠标快捷键
- 命令模式下双击cell任何部位——进入编辑模式
- 编辑模式下单击cell左侧——进入命令模式




## 键盘快捷键

### 命令模式：
- H：查看并设置所有快捷键
- Esc：关闭页面
- Q：关闭页面（有时不起作用）
- A：在上面插入cell
- B：在下面插入cell
- Shift-M：合并选中cell，如果只有一个cell被选中则默认与下面第一个cell合并，且合并后编辑环境与上面的代码快相同
- M：进入markdown编辑环境
- Y：进入code编辑环境
- R：清除代码块格式(环境变为非 code or markdown)
- 1~6：将代码块变为heading x
- L：显示/隐藏行号
- Shift+L：设置显示/隐藏行号并保存该设置
- O：展开/隐藏输出
- $\textrm{I I}$：中断服务
- 00：重启服务(带窗口)
- X：剪切选择的代码块
- C：复制选择的代码块
- Z：撤回
- V：粘贴
- DD：删除代码块
- Enter：进入编辑模式
- 双击cell：进入编辑模式




### 编辑模式：
- Tab：代码完成或缩进
- Shift-Tab：工具提示，如函数提示函数参数
- Ctrl+\]：缩进
- Ctrl+[：取消缩进

- Ctrl+D：删除整行
- Ctrl+U：撤销选择（感觉类似于撤销，还不知道有什么区别）
- Ctrl+Home：跳到单元格起始处
- Ctrl+End：跳到单元格最后
- Shift-Enter：运行代码块, 选择下面的代码块
- Ctrl-Enter：运行选中的代码块
- Alt-Enter：运行代码块并在下面插入新的cell
- Ctrl-Shift-Minus：在鼠标处分割代码块
- Ctrl-Shift-F/P：打开命令配置
- Ctrl+M：进入命令行模式
- Esc：进入命令行模式
- Insert：切换 重写标志（用了没反应
- Alt+U：重新选择（不知道，按了没反应）




## 远程访问 Jupyter

1. 在 shell 环境中生成配置文件（如果还未生成的话）

```bash
    jupyter notebook --generate-config
```
2. 打开 ipython，创建密码

```python
    In [1]: from notebook.auth import passwd
    In [2]: passwd()
    Enter password: 
    Verify password: 
    Out[2]: 'sha1:blabla...'
```

3. 在`C:/user/<user_name>/.jupyter/jupyter_notebook_config.py `中取消`c.NotebookApp.password = '...'`的注释，并将生成的秘钥复制进去，随后添加以下代码

```python
    c.NotebookApp.ip='*'  # 允许任何ip访问
    c.NotebookApp.port = 8888  # 自行指定一个端口, 访问时使用该端口
```

4. 远程访问时，若为服务器，则只需在浏览器中输入 `ip:8888` 类字样即可；若为私人设备，ip 可通过在源设备的 shell 环境打开 jupyter，便会出现类似以下字样，其中`http://LAPTOP-NDJ8L641:9695/`类字样即为所需地址

```bash
    [W 21:51:54.507 NotebookApp] WARNING: The notebook server is listening on all IP addresses and not using encryption. This is not recommended.
    [I 21:51:54.537 NotebookApp] [jupyter_nbextensions_configurator] enabled 0.4.1
    [I 21:51:54.537 NotebookApp] Serving notebooks from local directory: D:\Coding\Python\JupyterProjects
    [I 21:51:54.539 NotebookApp] Jupyter Notebook 6.1.1 is running at:
    [I 21:51:54.540 NotebookApp] http://LAPTOP-NDJ8L641:9695/
    [I 21:51:54.540 NotebookApp] Use Control-C to stop this server and shut down all kernels (twice to skip confirmation).
    [I 21:51:55.814 NotebookApp] 302 GET /tree (::1) 1.00ms
    [I 21:52:01.060 NotebookApp] 302 POST /login?next=%2Ftree (::1) 1.00ms
```

## jupyter配置

### 生成配置文件

```bash
    jupyter notebook --generate-config
```

### 设置默认浏览器为 chrome

在生成的`c:\\users\\<user_name>\\.jupyter\\jupyter_notebook_config.py`文件中找到`c.NotebookApp.browser`字段，将其取消注释并作如下修改：

```python
    import webbrowser

    webbrowser.register("chrome",None,webbrowser.GenericBrowser(u"C:\\...\\chrome.exe"))

    c.NotebookApp.browser = 'chrome'

```

### 设置jupyter默认打开的文件目录

找到`jupyter_notebook_config.py`文件（见上），找到`c.NotebookApp.notebook_dir`字段，将其取消注释并在后面加上希望的目录
